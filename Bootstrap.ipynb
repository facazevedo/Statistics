1. Bootstrap Sample Function: The function bootstrap_sample takes an array of data and the number of samples needed. It returns a sample generated by sampling with replacement from the given data.

2. Bootstrap Statistic Function: The function bootstrap_statistic generates the bootstrap distribution of a statistic (e.g., mean, median, standard deviation).

It takes the data, a statistic function, and the number of bootstrap samples to generate.

It initializes an array bootstrap_samples to store the bootstrap sample statistics.

It then generates num_bootstrap_samples bootstrap samples and calculates the statistic for each using the stat_function.

3. Sample Data: We have exam scores for a class stored in an array called data.

4. Statistic Function: We define the statistic function to be the mean (np.mean).

5. Generate Bootstrap Distribution: We call the bootstrap_statistic function to generate the bootstrap distribution of the mean.

6. Calculate Confidence Interval: We calculate the 95% confidence interval for the mean from the bootstrap distribution.

```python
import numpy as np

def bootstrap_sample(data, num_samples):
    """Generate a bootstrap sample from the given data."""
    return np.random.choice(data, size=num_samples, replace=True)

def bootstrap_statistic(data, stat_function, num_bootstrap_samples=10000):
    """Calculate the bootstrap distribution of a statistic."""
    bootstrap_samples = np.zeros(num_bootstrap_samples)
    
    for i in range(num_bootstrap_samples):
        sample = bootstrap_sample(data, len(data))
        bootstrap_samples[i] = stat_function(sample)
        
    return bootstrap_samples

# Sample data: Exam scores of a class
data = np.array([85, 88, 92, 95, 80, 73, 89, 90])

# Define a statistic function (here, mean)
stat_function = np.mean

# Generate bootstrap distribution of the mean
bootstrap_means = bootstrap_statistic(data, stat_function, num_bootstrap_samples=10000)

# Calculate 95% confidence interval
lower_bound = np.percentile(bootstrap_means, 2.5)
upper_bound = np.percentile(bootstrap_means, 97.5)

print(f"The 95% confidence interval for the mean is ({lower_bound}, {upper_bound})")
```
