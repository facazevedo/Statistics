# Types of distributions

##### Precise definitions: 

1. **Population**:
   The population is the entire group of individuals or objects that are the subject of a study. It encompasses all the members within a defined group.
   - **Example**: If you want to study the heights of adult males in a specific city, the population would be all adult males in that city.

2. **Attributes**:
   Attributes are specific characteristics or features that describe members of a population. They can be measured, observed, or categorized.
   - **Example**: In the study of the population of adult males in a city, the attribute of interest is their height.

3. **Population Distribution**:
   The population distribution refers to the distribution of a specific attribute (or range) within the population, not the distribution of the population itself. It describes how the values of that attribute are spread or dispersed across the population.
   - **Example**: The population distribution of heights among adult males in a city might show that 20% are between 5'5" and 5'8", 50% are between 5'9" and 6', and 30% are taller than 6'. This distribution tells you how the attribute of height is spread across the population of adult males in that city.
   
### Distribution of Attributes:
Refers to how specific characteristics are spread across a population, such as the distribution of heights among people in a city.

**Coin Flip Experiment**

Imagine you are flipping a coin twice, and you want to model the number of heads that appear. You can define a random variable X to represent this:

The **sample space (all possible outcomes)** is:
{
    (H, H),
    (H, T),
    (T, H),
    (T, T)
}

An **event** is a specific set of outcomes from a random experiment. It is a subset of the sample space
 
Examples:

- Single Outcome Event:
If you roll a six-sided die, the event "rolling a 3" consists of the single outcome {3}.

- Multiple Outcome Event:
The event "rolling an odd number" consists of the outcomes {1, 3, 5}.

- No Outcome Event: The event "rolling a 7 on a six-sided die" consists of no outcomes, since it's impossible.

The **random variable X** (function):
- X(H, H) = 2
- X(H, T) = 1
- X(T, H) = 1
- X(T, T) = 0

Here, X assigns a numerical value to each possible outcome, corresponding to the number of heads.

Basically, if a the random variable is the number of heads, the expected value of X would be = 1. Meaning that if I did the experiment many times, most of the times X=1 but not always. In algebra, the variable is deterministic. If x = 1, then x is always 1. The distribution of an algebraic variable has standard deviation = 0.

**Population**:

The population refers to the entire group that you're interested in studying. In this context, the population could be defined as all possible outcomes of flipping a coin twice. There are four possible outcomes in this population, each representing one possible combination of flips:

- HH (Head on the first flip and Head on the second flip)
- HT (Head on the first flip and Tail on the second flip)
- TH (Tail on the first flip and Head on the second flip)
- TT (Tail on the first flip and Tail on the second flip)

So the population consists of these four outcomes.

**Range (Set of Values)**:
{
    0,
    1,
    2
}

The range can be defined in relation to a specific random variable, which must be defined to give the range meaning.

For example, let's define a random variable X representing the number of Heads in two coin flips. In this case, the range of the random variable X would be the set of values it can take: {0, 1, 2}, corresponding to the number of Heads that can occur in two coin flips (0, 1, or 2).

**Probability Distribution**: Specifies the probability for each value in the range. It describes the behavior of the random variable:
- P(X = 0) = 1/4
- P(X = 1) = 1/2
- P(X = 2) = 1/4

### Difference:
- **Distribution of Attributes**: Descriptive summary of how attributes are spread in the real world (e.g., 20% of people are between 5'5" and 5'8").
- **Distribution of Random Variables**: Mathematical model of the likelihoods of different outcomes (e.g., probability that \(X\), representing height, falls within a specific range).

### Relationship:
The distribution of a random variable related to an attribute (e.g., height) can be seen as a mathematical representation of the distribution of that attribute within the population.

## Discrete distributions:

#### 1. Binomial Distribution
**Definition**: The binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success.

$$ P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} $$
where:
- $ n $ is the number of trials.
- $ k $ is the number of successes.
- $ p $ is the probability of success on an individual trial.

**Example**: Imagine flipping a coin 10 times. The number of heads (successes) you might get, say exactly 6 heads, can be described by the binomial distribution.

---

#### 2. Poisson Distribution
**Definition**: The Poisson distribution gives the probability of a given number of events happening in a fixed interval of time or space, given the average number of times the event occurs over that period.
 
$$ P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!} $$
where:
- $ k $ is the actual number of successes that result from the experiment.
- $ \lambda $ is the mean number of successes that occur in a specified region.

**Example**: If an average of 7 cars pass through a traffic signal in 10 minutes, the probability of seeing a different number of cars, say exactly 10 cars, in a randomly selected 10-minute interval is modeled by the Poisson distribution.

---

#### 3. Geometric Distribution
**Definition**: The geometric distribution provides the probability of the number of Bernoulli trials required for the first success to occur.

$$ P(X = k) = (1-p)^{k-1} p $$
where:
- $ k $ is the number of trials needed for the first success.
- $ p $ is the probability of success on an individual trial.

**Example**: If you keep flipping a coin, the number of flips you might need to get the first head can be described by the geometric distribution.

---

#### 4. Bernoulli Distribution
**Definition**: The Bernoulli distribution represents an experiment where the outcome can be classified as either a success or a failure.

$$ P(X = k) = \begin{cases} 
p & \text{if } k = 1 \\
1-p & \text{if } k = 0 
\end{cases} $$
where:
- $ k $ is the outcome of the experiment, either 0 (failure) or 1 (success).
- $ p $ is the probability of success.

**Example**: The outcome of a single coin flip can be either a head (success) or a tail (failure), and this is a classic example of a Bernoulli distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, poisson, geom, bernoulli
from matplotlib.lines import Line2D

np.random.seed(0)

# Data Generation
data_binom = binom.rvs(10, 0.5, size=10000)
lambda_val = 7
data_poisson = poisson.rvs(lambda_val, size=10000)
p = 0.5
data_geom = geom.rvs(p, size=10000)
data_bernoulli = bernoulli.rvs(0.5, size=10000)

# Plotting
fig, ax = plt.subplots(1, 4, figsize=(20, 6))

# Binomial Distribution
sns.histplot(data_binom, bins=20, kde=False, ax=ax[0], stat="probability", discrete=True, shrink=0.5)
ax[0].set_title('Binomial Distribution\n(Number of heads in 10 coin flips)')
x = np.arange(0, 11)
ax[0].plot(x, binom.pmf(x, 10, 0.5), 'o-', color='red') # Theoretical representation

# Poisson Distribution
sns.histplot(data_poisson, bins=40, kde=False, ax=ax[1], stat="probability", discrete=True, shrink=0.5)
ax[1].set_title('Poisson Distribution\n(Cars passing through a traffic signal in 10 minutes)')
x = np.arange(0, 3*lambda_val)
ax[1].plot(x, poisson.pmf(x, lambda_val), 'o-', color='red') # Theoretical representation

# Geometric Distribution
sns.histplot(data_geom, bins=range(1, 15), kde=False, ax=ax[2], stat="probability", discrete=True, shrink=0.5)
ax[2].set_title('Geometric Distribution\n(Tosses to get the first head)')
x = np.arange(1, 15)
ax[2].plot(x, geom.pmf(x, p), 'o-', color='red') # Theoretical representation

# Bernoulli Distribution
sns.histplot(data_bernoulli, bins=3, kde=False, ax=ax[3], stat="probability", discrete=True, shrink=0.5)
ax[3].set_title('Bernoulli Distribution\n(Single coin flip)')
x = [0, 1]
ax[3].plot(x, bernoulli.pmf(x, 0.5), 'o-', color='red') # Theoretical representation

legend_elements = [Line2D([0], [0], marker='o', color='w', label='Probability Mass Function (PMF)', markersize=10, markerfacecolor='red')]
fig.legend(handles=legend_elements, loc='upper center', ncol=4, bbox_to_anchor=(0.5, 1.15))

fig.subplots_adjust(top=0.75)
plt.tight_layout()
plt.show()
```

## Continuous distributions:

#### 1. Normal (Gaussian) Distribution
**Definition**: The normal (or Gaussian) distribution describes a symmetric bell-shaped distribution that is fully defined by its mean and standard deviation.

$$ f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{1}{2} \left(\frac{x-\mu}{\sigma}\right)^2 } $$
where:
- $ \mu $ is the mean.
- $ \sigma $ is the standard deviation.

**Example**: Heights of adult humans. Most people have heights around the average (mean) height, and fewer are very short or very tall.

---

#### 2. Exponential Distribution
**Definition**: The exponential distribution describes the time intervals in a homogeneous Poisson process.

$$ f(x|\lambda) = \lambda e^{-\lambda x} $$
for $ x \geq 0 $ and $ \lambda > 0 $

**Example**: The time between events in a Poisson process, such as the time taken for the next customer to arrive at a service center.

---

#### 3. Uniform Distribution
**Definition**: The uniform distribution ensures that all values have an equal chance of being selected. The distribution is defined by two parameters, a and b, which are the minimum and maximum values.

$$ f(x) = \frac{1}{b-a} $$
for $ a \leq x \leq b $

**Example**: Choosing a random number between 0 and 1. Each possible value in this interval has an equal chance of being chosen.

---

#### 4. Beta Distribution
**Definition**: The beta distribution describes a distribution on the interval [0,1], which is useful in Bayesian statistics and is determined by two shape parameters, $\alpha$ and $\beta$.

$$ f(x;\alpha,\beta) = \frac{x^{\alpha-1} (1-x)^{\beta-1}}{B(\alpha,\beta)} $$
for $ 0 \leq x \leq 1 $ and $ \alpha, \beta > 0 $

**Example**: The distribution of probabilities in Bayesian statistics, such as estimating the probability of success in a binomial experiment.

---

#### 5. Chi-squared Distribution
**Definition**: The chi-squared distribution is used in statistical inference and describes the distribution of the sum of squared standard normal variables.

$$ f(x; k) = \frac{1}{2^{k/2} \Gamma(k/2)} x^{k/2 - 1} e^{-x/2} $$
for $ x > 0 $ and $ k > 0 $

**Example**: Used in hypothesis testing, especially for tests of goodness of fit.

---

#### 6. F Distribution
**Definition**: The F distribution describes the distribution of the ratio of two independent chi-squared random variables, each divided by their degrees of freedom.

**Example**: Used in ANOVA (analysis of variance) and regression analysis.

---

#### 7. Student's t Distribution
**Definition**: The Student's t distribution is like the normal distribution but has heavier tails. It is used when the sample size is small or the population standard deviation is unknown.

$$ f(t; v) = \frac{1}{\sqrt{v} B(v/2, 1/2)} (1 + t^2/v)^{-\frac{v+1}{2}} $$
for all real numbers $ t $

**Example**: Used in hypothesis testing and constructing confidence intervals, especially when the sample size is small.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, uniform, beta, chi2, f, t

np.random.seed(0)

# Plotting
fig, ax = plt.subplots(2, 4, figsize=(20, 12))

# Function to add dashed vertical line at the mode
def add_dashed_peak_line(ax, x_range, pdf_values):
    x_peak = x_range[np.argmax(pdf_values)]
    ax.axvline(x_peak, linestyle='--', color='black')

# Normal Distribution
x = np.linspace(-4, 4, 100)
pdf_values = norm.pdf(x, loc=0, scale=1)
ax[0,0].plot(x, pdf_values, '-', color='red')
add_dashed_peak_line(ax[0,0], x, pdf_values)
ax[0,0].set_title('Normal Distribution\n(Mean=0, Std Deviation=1)')
ax[0,0].set_ylabel('Density')

# Exponential Distribution
x = np.linspace(0, 8, 100)
pdf_values = expon.pdf(x, scale=1)
ax[0,1].plot(x, pdf_values, '-', color='red')
add_dashed_peak_line(ax[0,1], x, pdf_values)
ax[0,1].set_title('Exponential Distribution\n(Scale=1)')

# Uniform Distribution
x = np.linspace(0, 1, 100)
pdf_values = uniform.pdf(x, loc=0, scale=1)
ax[0,2].plot(x, pdf_values, '-', color='red')
# No peak line for uniform since it's flat
ax[0,2].set_title('Uniform Distribution\n(0 to 1)')

# Beta Distribution
x = np.linspace(0, 1, 100)
pdf_values = beta.pdf(x, a=2, b=5)
ax[0,3].plot(x, pdf_values, '-', color='red')
add_dashed_peak_line(ax[0,3], x, pdf_values)
ax[0,3].set_title('Beta Distribution\n(a=2, b=5)')

# Chi-squared Distribution
x = np.linspace(0, 20, 100)
pdf_values = chi2.pdf(x, df=4)
ax[1,0].plot(x, pdf_values, '-', color='red')
add_dashed_peak_line(ax[1,0], x, pdf_values)
ax[1,0].set_title("Chi-squared Distribution\n(df=4)")
ax[1,0].set_ylabel('Density')
ax[1,0].set_xlabel('x')

# F Distribution
x = np.linspace(0, 5, 100)
pdf_values = f.pdf(x, 5, 2)
ax[1,1].plot(x, pdf_values, '-', color='red')
add_dashed_peak_line(ax[1,1], x, pdf_values)
ax[1,1].set_title("F Distribution\n(df1=5, df2=2)")
ax[1,1].set_xlabel('x')

# Student's t Distribution
x = np.linspace(-5, 5, 100)
pdf_values = t.pdf(x, df=10)
ax[1,2].plot(x, pdf_values, '-', color='red')
add_dashed_peak_line(ax[1,2], x, pdf_values)
ax[1,2].set_title("Student's t Distribution\n(df=10)")
ax[1,2].set_xlabel('x')

# Leave last plot empty (for symmetry)
ax[1,3].axis('off')

fig.subplots_adjust(top=0.9, hspace=0.5)
plt.tight_layout()
plt.show()
```

### Gaussians

$$ f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{1}{2} \left(\frac{x-\mu}{\sigma}\right)^2 } $$
where:
- $ \mu $ is the mean.
- $ \sigma $ is the standard deviation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for Gaussian curves
num_curves = 5
mu_values = [0, 1, 2, -1, -2]
sigma_values = [0.5, 1, 1.5, 2, 2.5]

# Generate x values
x = np.linspace(-5, 5, 500)

# Create a figure with subplots
plt.figure(figsize=(15, 6))

# Plot Gaussian curves with different means
plt.subplot(1, 2, 1)
for mu in mu_values:
    y = 1 / (1 * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mu) / 1)**2)
    label = f'Mean = {mu}'
    plt.plot(x, y, label=label)

plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Gaussian Curves with Different Means')
plt.legend()
plt.grid(True)

# Plot Gaussian curves with different standard deviations
plt.subplot(1, 2, 2)
for sigma in sigma_values:
    y = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - 0) / sigma)**2)
    label = f'Standard Deviation = {sigma}'
    plt.plot(x, y, label=label)

plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Gaussian Curves with Different Standard Deviations')
plt.legend()
plt.grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
