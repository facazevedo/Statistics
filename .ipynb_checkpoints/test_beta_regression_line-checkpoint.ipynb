## Step 3: Hypothesis Testing

### Hypothesis Statement

- **Null Hypothesis ($ H_0 $)**: The slope is zero ($ \beta_1 = 0 $), implying that the independent variable has no effect on the dependent variable.
- **Alternative Hypothesis ($ H_a $)**: The slope is not zero ($ \beta_1 \neq 0 $), implying that the independent variable has a significant effect on the dependent variable.

### Calculation of Residuals

Residuals are the differences between the observed values ($ Y $) and the values predicted by the model ($ \hat{Y} $). The residuals are calculated as:

$$
\text{residuals} = Y - \hat{Y}
$$

### Standard Error of the Residuals

The standard error (stderr) of the residuals is calculated as the standard deviation of the residuals. This measure helps us understand the average variation of the residual values from the fitted line.

### T-Statistic

The t-statistic for the slope is calculated using the formula:

$$
t = \frac{\beta_1}{\text{SE}(\beta_1)}
$$

Where $ \text{SE}(\beta_1) $ is the standard error of the slope $ \beta_1 $, calculated as:

$$
\text{SE}(\beta_1) = \frac{\text{stderr}}{\sqrt{\sum{(X - \bar{X})^2}}}
$$

### Degrees of Freedom

The degrees of freedom for this test are $ n - 2 $, where $ n $ is the number of observations. The "-2" accounts for the two parameters estimated in the model: the intercept ($ \beta_0 $) and the slope ($ \beta_1 $).

### P-Value

The p-value is calculated based on the t-statistic and the degrees of freedom. The p-value gives us the probability of observing such a t-statistic if the null hypothesis is true. The two-tailed p-value is used in this case because the alternative hypothesis is $ \beta_1 \neq 0 $.

### Significance Testing

Finally, we compare the p-value to a predetermined significance level ($ \alpha $, usually 0.05). If the p-value is less than $ \alpha $, we reject the null hypothesis, concluding that $ \beta_1 $ is statistically different from zero.

- $ p \text{-value} < \alpha $: Reject the null hypothesis. The slope is statistically different from zero.
- $ p \text{-value} \geq \alpha $: Fail to reject the null hypothesis. The slope is not statistically different from zero.

This hypothesis test gives you statistical confidence that the independent variable is (or is not) a significant predictor of the dependent variable.


```python
# Import required libraries
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Generate synthetic data
np.random.seed(42)  # For reproducibility

# Generate 100 data points for the independent variable (X)
X = np.linspace(0, 10, 100)

# Generate the dependent variable (Y) with some noise
# Actual equation: Y = 2 * X + 1 + noise
noise = np.random.normal(0, 1, 100)
Y = 2 * X + 1 + noise

# Plot the synthetic data
plt.scatter(X, Y)
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.title('Synthetic Data')
plt.show()

# Step 2: Fit a simple linear regression model
X_reshaped = X.reshape(-1, 1)  # Reshape for sklearn
model = LinearRegression()
model.fit(X_reshaped, Y)

# Extract slope (beta_1) and intercept (beta_0)
beta_1 = model.coef_[0]
beta_0 = model.intercept_
print(f"Intercept (beta_0): {beta_0}")
print(f"Slope (beta_1): {beta_1}")

# Step 3: Hypothesis Testing
# Calculate residuals
residuals = Y - model.predict(X_reshaped)

# Calculate standard error of the residuals
stderr = np.std(residuals)

# Perform t-test to check if slope is statistically different from zero
# Null Hypothesis: beta_1 = 0
# Alternative Hypothesis: beta_1 != 0

# Calculate t-statistic
t_statistic = beta_1 / (stderr / np.sqrt(np.sum((X - np.mean(X)) ** 2)))

# Calculate degrees of freedom
degrees_of_freedom = len(X) - 2  # n - 2, where n is number of observations

# Calculate p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), degrees_of_freedom))

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Evaluate the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis. The slope is statistically different from zero.")
else:
    print("Fail to reject the null hypothesis. The slope is not statistically different from zero.")
```
