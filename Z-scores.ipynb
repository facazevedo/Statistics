# Conversion to Z-Scores

## Definition:

The **z-score** of an observation is defined as:

$$ z = \frac{x - \mu}{\sigma} $$

Where:
- $ x $ is the raw score (or value) to be standardized.
- $ \mu $ is the mean of the population or sample from which $ x $ is drawn.
- $ \sigma $ is the standard deviation of the population or sample.

## What Does It Do?

1. **Centering**: A z-score transformation recenters the data around zero. After conversion, the mean of the z-scores will always be 0.
   
$$ \mu_{z} = \mu_{x} - \frac{\mu_{x}}{\sigma_{x}} = 0 $$

2. **Scaling**: A z-score transformation rescales the data so that the variance (and standard deviation) is 1. In the case of a 2D plot, where y = count and x = feature, the scaling only happens along the X axis.

$$ \sigma^{2}_{z} = \frac{1}{n} \sum_{i=1}^{n} (z_{i} - \mu_{z})^2 = \frac{1}{n} \sum_{i=1}^{n} \left(\frac{x_{i} - \mu_{x}}{\sigma_{x}}\right)^2 = 1 $$

## Advantages:

1. **Standardization**: By converting data into z-scores, you're essentially standardizing your data to a mean of 0 and a standard deviation of 1. This makes it easier to compare scores from different distributions, since they now share a common scale.

2. **Comparison Across Different Units**: If two sets of data are measured in different units (e.g., height in inches vs. centimeters), converting to z-scores can put them on a common scale, making them directly comparable.

3. **Detecting Outliers**: Observations with large absolute z-scores can be considered outliers. A common rule of thumb, for instance, is to consider observations with z-scores greater than 2 or less than -2 as potential outliers, although the exact threshold can vary.

4. **Facilitates Statistical Analysis**: Many statistical techniques assume data is normally distributed. By working with z-scores, it becomes easier to assess the position of an observation within the context of a normal distribution. This is useful in hypothesis testing and in various methods in inferential statistics.

5. **Understanding Data Position**: It provides a way to understand where a particular score lies in relation to the rest. A z-score tells you how many standard deviations away from the mean your observation is.

## How Does It Change the Data?

1. **No Change in Shape**: The transformation doesn't affect the shape of the distribution of the data. If the original data is normally distributed, the z-scores will also be normally distributed.

2. **Change in Location and Scale**: The location (mean) and scale (standard deviation) of the distribution are changed to 0 and 1, respectively.

### Setup

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Dice PMF
def f_dice(num_sides_dice):
    pmf_values = 1/num_sides_dice
    return pmf_values

# Human height
def f_height(x):
    mu = 70  # mean
    sigma = 3  # standard deviation
    pdf_values = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    return pdf_values

# Lightbulb lifespan
def f_lightbulb(x):
    lambda_ = 1/10
    pdf_values = lambda_ * np.exp(-lambda_ * x)
    return pdf_values

def discretize_curve(f, start, end, n_bins, total_count):
    bin_edges = np.linspace(start, end, n_bins + 1)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2.0
    bin_width = (end - start) / n_bins
    densities = np.array([np.mean(f(np.linspace(edge, bin_edges[i+1], 100))) for i, edge in enumerate(bin_edges[:-1])])
    float_counts = densities * total_count / sum(densities)
    int_counts = np.round(float_counts).astype(int)

    # Remove elements where int_counts is 0
    non_zero_indices = int_counts != 0
    bin_centers = bin_centers[non_zero_indices]
    int_counts = int_counts[non_zero_indices]

    return bin_centers, int_counts

def discretize_dice(num_sides_dice, total_count):
    pmf_value = f_dice(num_sides_dice)
    counts = [int(pmf_value * total_count) for _ in range(num_sides_dice)]
    
    # Distribute any remaining counts
    remainder = total_count - sum(counts)
    for i in range(remainder):
        counts[i % num_sides_dice] += 1
    
    # Remove elements where counts is 0
    non_zero_indices = np.array(counts) != 0
    centers = np.arange(1, num_sides_dice + 1)[non_zero_indices]
    counts = np.array(counts)[non_zero_indices].tolist()

    return centers, counts

def plot_discretized_distribution(ax, centers, counts, title, xlabel, ylabel, bin_width=None, is_dice=False):
    # If bin_width is not provided, then calculate it dynamically
    if bin_width is None:
        if len(centers) > 1:
            bin_width = min(np.diff(centers))
        else:
            bin_width = 1
    
    ax.bar(centers, counts, width=bin_width, alpha=0.5, edgecolor='k')
    ax.set_xticks(centers)
    if len(centers) > 12:
        ax.tick_params(axis='x', rotation=90)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)

### MAIN ###

# Similar to before, here we define the number of distinct elements our sample can have in the discretized population. The discretized pop. should follow the pop. curve as close as possible.
# First we calculate the max amount of uniformly distributed bins our subpopulation can have.
num_bins_dice = 6
num_bins_height = 100
num_bins_lightbulb = 100

# Then we define the size of our subpopulation
total_count_dice = 100
total_count_height = 100
total_count_lightbulb = 100

# Discretize dice
centers_dice, counts_dice = discretize_dice(num_bins_dice, total_count_dice)
#print(f"\nDice\nNumber of bins: {len(counts_dice)}\nCounts: {sum(counts_dice)}")
df_dice = pd.DataFrame({'X': centers_dice, 'counts': counts_dice})
#display(df_dice)

# Discretize height
centers_height, counts_height = discretize_curve(f_height, 55, 85, num_bins_height, total_count_height)
#print(f"\nHuman heights\nNumber of bins: {len(counts_height)}\nCounts: {sum(counts_height)}")
df_height = pd.DataFrame({'X': centers_height, 'counts': counts_height})
#display(df_height)

# Discretize lightbulb
centers_lightbulb, counts_lightbulb = discretize_curve(f_lightbulb, 0, 50, num_bins_lightbulb, total_count_lightbulb)
#print(f"\nLightbulb lifespan\nNumber of bins: {len(counts_lightbulb)}\nCounts: {sum(counts_lightbulb)}")
df_lightbulb = pd.DataFrame({'X': centers_lightbulb, 'counts': counts_lightbulb})
#display(df_lightbulb)
    
# Plot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
# plot_discretized_distribution(axes[0], centers_dice, counts_dice, 1, "Discretized population of dice rolls", "Dice Value", "Count", is_dice=True)
# plot_discretized_distribution(axes[1], centers_height, counts_height, (85-55)/num_bins_height, "Discretized population of human heights", "Height (inches)", "Count")
# plot_discretized_distribution(axes[2], centers_lightbulb, counts_lightbulb, 50/num_bins_lightbulb, "Discretized population of lightbulb fifespans", "Lifespan (units)", "Count")

plot_discretized_distribution(axes[0], centers_dice, counts_dice, "Discretized population of dice rolls", "Dice Value", "Count", 1, is_dice=True)
plot_discretized_distribution(axes[1], centers_height, counts_height, "Discretized population of human heights", "Height (inches)", "Count", (85-55)/num_bins_height)
plot_discretized_distribution(axes[2], centers_lightbulb, counts_lightbulb, "Discretized population of lightbulb fifespans", "Lifespan (units)", "Count", 50/num_bins_lightbulb)

plt.tight_layout()
plt.show()
```

#### Convert to z-scores

```python
def calculate_z_scores(centers, counts):
    # Calculate the mean and standard deviation
    mean = sum(centers * counts) / sum(counts)
    variance = sum(counts * (centers - mean)**2) / sum(counts)
    std_dev = np.sqrt(variance)
    
    # Calculate z-scores
    z_scores = (centers - mean) / std_dev
    
    return z_scores

# Calculate z-scores for the three distributions
z_scores_dice = calculate_z_scores(centers_dice, counts_dice)
z_scores_height = calculate_z_scores(centers_height, counts_height)
z_scores_lightbulb = calculate_z_scores(centers_lightbulb, counts_lightbulb)

# Plot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

plot_discretized_distribution(axes[0], z_scores_dice, counts_dice, "Z-scores for Dice Rolls", "Z-score", "Count", is_dice=True)
plot_discretized_distribution(axes[1], z_scores_height, counts_height, "Z-scores for Human Heights", "Z-score", "Count")
plot_discretized_distribution(axes[2], z_scores_lightbulb, counts_lightbulb, "Z-scores for Lightbulbs Lifespans", "Z-score", "Count")

plt.tight_layout()
plt.show()
```
